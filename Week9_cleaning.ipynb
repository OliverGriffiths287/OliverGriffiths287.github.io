```python
import pandas as pd
import csv

with open('/content/prices_takeaways.csv', 'r') as file:
    reader = csv.reader(file)
    data = list(reader)

df = pd.DataFrame(data[1:], columns=data[0])
df = df[df['item_id'] == '220319']
df.loc[df['region'] == '2', 'region'] = 'London'
df.loc[df['region'] == '3', 'region'] = 'South East (England)'
df.loc[df['region'] == '4', 'region'] = 'South West (England)'
df.loc[df['region'] == '5', 'region'] = 'East'
df.loc[df['region'] == '6', 'region'] = 'East Midlands (England)'
df.loc[df['region'] == '7', 'region'] = 'West Midlands (England)'
df.loc[df['region'] == '8', 'region'] = 'Yorkshire and The Humber'
df.loc[df['region'] == '9', 'region'] = 'North West (England)'
df.loc[df['region'] == '10', 'region'] = 'North East (England)'
df.loc[df['region'] == '11', 'region'] = 'Wales'
df.loc[df['region'] == '12', 'region'] = 'Scotland'
df.loc[df['region'] == '13', 'region'] = 'Northern Ireland'
df['price'] = pd.to_numeric(df['price'], errors='coerce')
df['quote_date'] = pd.to_datetime(df['quote_date'], format= '%Y%m').dt.strftime('%Y-%m')
df = df.rename(columns={'quote_date': 'month'})
df['year'] = pd.to_datetime(df['month'], format= '%Y-%m').dt.strftime('%Y')

print(df.dtypes)

df
```
