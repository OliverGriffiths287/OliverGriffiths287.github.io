The first half of code was an original idea for a potential chart looking at where Premier League owners are from, although the second half is actually used in my project.

```python
import requests
import re
import json
import pandas as pd
```

```python
from mmap import PAGESIZE
url = "https://en.wikipedia.org/wiki/List_of_owners_of_English_football_clubs"

req = requests.get(url)
page_src1 = req.text

start_index = page_src1.index('<div class="mw-heading mw-heading2"><h2 id="Premier_League"')
page_src1 = page_src1[start_index:]

page_src1 = page_src1.split('title="Fosun International">')[0]


page_src1
```

```python
from os import name
# Find all team entries using regex
nation_entries = re.findall(r'"><img alt="(.*?)" src="', page_src1)
nation_entries
```

```python
dfowners = pd.DataFrame({'Country': nation_entries})
result = dfowners['Country'].value_counts().reset_index()
result.columns = ['Country', 'Count']
result

result.to_csv('owners24.csv', index=False)
```
