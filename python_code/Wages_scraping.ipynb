```python
import requests
import re
import json
import pandas as pd
```

```python
from mmap import PAGESIZE
url = "https://www.planetfootball.com/quick-reads/premier-league-wages-increase-figures"

req = requests.get(url)
page_src = req.text

start_index = page_src.index('/><strong>')
page_src = page_src[start_index:]

page_src = page_src.split('Brighton')[0]


page_src

```

```python
from os import name
# Find all team entries using regex
team_entries = re.findall(r'<strong>(.*?)</strong>', page_src)

cleaned_teams = [name.replace('\xa0', '') for name in team_entries]

teams = cleaned_teams[:9] + [cleaned_teams[11]]

teams

dfteams = pd.DataFrame(teams)
dfteams.rename(columns={0: 'team'}, inplace=True)
dfteams
```

```python
# For all 2013-14 wages
wages_14 = re.findall(r'2013-14 Wages: (.*?)<br', page_src)
wages_13 = re.findall(r'2013-14 Wages:\xa0(.*?)<br', page_src)

wages_14 = wages_14 + wages_13
wages_14

dfwages_14 = pd.DataFrame(wages_14)
dfwages_14
dfwages_14.rename(columns={0: 'wages'}, inplace=True)
dfwages_14['season'] = '2014'
dfwages_14.loc[dfwages_14['wages'] == '77,784', 'team'] = 'Manchester City'
dfwages_14.loc[dfwages_14['wages'] == '50,568', 'team'] = 'Manchester United'
dfwages_14.loc[dfwages_14['wages'] == '40,393', 'team'] = 'Chelsea'
dfwages_14.loc[dfwages_14['wages'] == '36,070', 'team'] = 'Liverpool'
dfwages_14.loc[dfwages_14['wages'] == '33,815', 'team'] = 'Arsenal'
dfwages_14.loc[dfwages_14['wages'] == '24,039', 'team'] = 'Tottenham'
dfwages_14.loc[dfwages_14['wages'] == '17,514', 'team'] = 'Aston Villa'
dfwages_14.loc[dfwages_14['wages'] == '20,200', 'team'] = 'West Ham'
dfwages_14.loc[dfwages_14['wages'] == '29,033', 'team'] = 'Everton'
dfwages_14.loc[dfwages_14['wages'] == '8281', 'team'] = 'Crystal Palace'
dfwages_14.loc[dfwages_14['wages'] == '18,930', 'team'] = 'Newcastle United'
dfwages_14['wages'] = pd.to_numeric(dfwages_14['wages'].str.replace(',', ''), errors='coerce')
dfwages_14.reset_index(drop=True, inplace=True)
dfwages_14.drop(dfwages_14[dfwages_14['team'].isnull()].index, inplace=True)
dfwages_14

```

```python
wages_23 = re.findall(r'2022-23 Wages: (.*?)<br', page_src)
wages_22 = re.findall(r'2022-23 Wages:\xa0(.*?)<br', page_src)

wages_23 = wages_23 + wages_22

dfwages_23 = pd.DataFrame(wages_23)
dfwages_23
dfwages_23.rename(columns={0: 'wages'}, inplace=True)
dfwages_23['season'] = '2023'
dfwages_23.loc[dfwages_23['wages'] == '146,875', 'team'] = 'Manchester City'
dfwages_23.loc[dfwages_23['wages'] == '133,448', 'team'] = 'Manchester United'
dfwages_23.loc[dfwages_23['wages'] == '124,242', 'team'] = 'Chelsea'
dfwages_23.loc[dfwages_23['wages'] == '104,167', 'team'] = 'Liverpool'
dfwages_23.loc[dfwages_23['wages'] == '83,609', 'team'] = 'Arsenal'
dfwages_23.loc[dfwages_23['wages'] == '80,385', 'team'] = 'Tottenham'
dfwages_23.loc[dfwages_23['wages'] == '75,995', 'team'] = 'Aston Villa'
dfwages_23.loc[dfwages_23['wages'] == '68,182', 'team'] = 'West Ham'
dfwages_23.loc[dfwages_23['wages'] == '61,440', 'team'] = 'Everton'
dfwages_23.loc[dfwages_23['wages'] == '53,750', 'team'] = 'Crystal Palace'
dfwages_23.loc[dfwages_23['wages'] == '50,017', 'team'] = 'Newcastle United'
dfwages_23['wages'] = pd.to_numeric(dfwages_23['wages'].str.replace(',', ''), errors='coerce')
dfwages_23.reset_index(drop=True, inplace=True)
dfwages_23.drop(dfwages_23[dfwages_23['team'].isnull()].index, inplace=True)
dfwages_23
```

```python
dfwage_14_23 = pd.concat([dfwages_14, dfwages_23], ignore_index=True)
dfwage_14_23

dfwage_14_23.to_csv('wages_14_23.csv', index=False)
```
